USE-CASE 2 : use association learning to display shows similar to what user is about to watch.

```python
netflix_data = pd.read_csv("../cleaned_netflix_data.csv")
netflix_data.head()
```

```python
viewers_data = pd.read_csv("../viewers.csv")
viewers_data.head()
```

```python
len(viewers_data.title.unique())
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
# create dataframe with username and show titles watched by the user
user_show_df = pd.DataFrame(columns=['username', 'show_title']) 


# iterate through unique usernames
for username in viewers_data['username'].unique():
    
    # get the shows watched by the user
    shows_watched = viewers_data[viewers_data['username'] == username]['title'].values.tolist()

    # add the username and shows watched by the user to the dataframe
    user_show_df = user_show_df.append({'username': username, 'show_title': shows_watched}, ignore_index=True)
```

```python
user_show_df.head()
```

```python
data = list(user_show_df['show_title'].values)
data[:3]
```

Creating Apriori Data Structure

```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_data = te.fit(data).transform(data).astype(int)
df = pd.DataFrame(te_data, columns=te.columns_)
df.head()
```

```python
df.shape
```

```python
freq_data = apriori(df, min_support=0.0005, use_colnames=True, verbose=1, max_len=2, low_memory=True)
freq_data
```

```python
rules = association_rules(freq_data, metric="lift", min_threshold=0.1)
sorted = rules.sort_values(by='lift', ascending=False)
sorted
```

```python
sorted.shape
```

Recommend shows based on given show

```python
def recommend_shows(showName):
  recommend_shows = []
  for i, show in sorted["antecedents"].items():
    for j in list(show):
        if j == showName:
            recommend_shows.append(list(sorted.iloc[i]["consequents"]))

  recommend_shows = list({item for item_list in recommend_shows for item in item_list}) # To get unique products
  return recommend_shows[:5]
```

```python
import random
recommend_shows(random.choice(netflix_data.title))
```
